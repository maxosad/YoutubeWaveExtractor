```python
import numpy as np

import matplotlib.pyplot as plt

import matplotlib.patches as patches
from matplotlib.path import Path
```

```python
def str_to_float_cortage(string_two_floats):
    (s1, s2) = string_two_floats.split(',')
    return (float(s1), float(s2))
```

```python
# 0 100 э то превоначальный отступ. ось y направлена вниз
```

```python
verts = [
   (0., 100.)   # P0
]

codes = [
    Path.MOVETO
]

for i in range(len(biz)):
    if (biz[i] == 'M'):
        i = i + 1
    elif (biz[i] == 'C'):
        for k in range(3):
            i = i + 1
            verts.append(str_to_float_cortage(biz[i]))
            codes.append(Path.CURVE3)
```

```python
path = Path(verts, codes)

fig, ax = plt.subplots()
patch = patches.PathPatch(path, facecolor='none', lw=2)
ax.add_patch(patch)

xs, ys = zip(*verts)
plt.gca().invert_yaxis ()
ax.plot(xs, ys,  lw=1, color='black', ms=10)
```
